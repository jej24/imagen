***CARGAR IMAGEN ORIGINAL Y VISUALIZAR CON MATPLOTLIB***

```
#leer una imagen

img_BGR = cv2.imread('/content/drive/My Drive/PDI/FOTO/prueba1.jpg', cv2.IMREAD_COLOR)   #antes era imread_unchanged ahora es imread_color

plt.figure(figsize=(7,7))    #(7,7) el tamaño del frame NO DE LA IMAGEN
plt.imshow(img_BGR)
plt.show()
```

***CAMBIAR TAMAÑO DE LA IMAGEN***

```
#expandir

print(img.shape)
res = cv2.resize(img, None, fx=10, fy=10, interpolation = cv2.INTER_CUBIC)                #interpolation adicional valores que no conocia  #res es para guardar el resultado, aai se nombró la variable
print(res.shape)
plt.imshow(res)
plt.show()


## sale (300, 300, 3)  sale 3 porque es a color, la matriz de color
```

***CONTRAER UNA IMAGEN***

```
#SHRINKING (contraer)

print(img.shape)
res = cv2.resize(img, None, fx=.1, fy=.1, interpolation = cv2.INTER_CUBIC)                #interpolation adicional valores que no conocia  #res es para guardar el resultado, aai se nombró la variable
print(res.shape)
plt.imshow(res)
plt.show()
```

***REMOVER LOS NUMEROS***

```
plt.figure(figsize=(7,7))    #(7,7) el tamaño del frame NO DE LA IMAGEN
plt.imshow(img_BGR)
plt.xticks([])
plt.yticks([])
plt.show()
```

***CAMBIAR BGR A RGB***

```
#pasar la imagen a RGB

img_RGB= cv2.cvtColor(img_BGR, cv2.COLOR_BGR2RGB)  #!!!!!!  DIFERENCIAR ENTRE MAYUSCULAS Y MINUSCULASS  !!!!!!

plt.figure(figsize=(10,10))    
plt.imshow(img_RGB)
plt.xticks([])
plt.yticks([])
plt.show()
```
